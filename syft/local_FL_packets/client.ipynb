```python
in_dim = 188
out_dim = 2
n_samples = 150
```

```python
class Network(nn.Module):
  def __init__(self):
    super(Network, self).__init__()
    self.layer1 = nn.Linear(in_dim, 150)
    self.layer2 = nn.Linear(150, 50)
    self.layer3 = nn.Linear(50, 30)
    self.out = nn.Linear(30, out_dim)
    
  def forward(self, x):
        x = nn.functional.relu(self.layer1(x))
        x = nn.functional.relu(self.layer2(x))
        x = nn.functional.relu(self.layer3(x))
        output = nn.functional.log_softmax(self.out(x), dim=1)
        return output
```

```python
clf = Network()
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = th.optim.Adam(clf.parameters(), lr=0.1)
```

```python
def train(iterations, model, optim, data_ptr, target_ptr):

    losses = []

    for i in range(iterations):

        optim.zero_grad()

        output = model(data_ptr)

        # nll_loss = negative log-liklihood loss
        loss = nn.functional.nll_loss(output, target_ptr.long())

        loss_item = loss.item()

        loss_value = loss_item.get(
            reason="To evaluate training progress", request_block=True, timeout_secs=5
        )

        if i % 10 == 0:
            print("Epoch", i, "loss", loss_value)

        losses.append(loss_value)

        loss.backward()

        optim.step()

    return losses
```

```python
iteration = 50
losses = train(iteration, clf, optimizer, tt_data_X, tt_data_y)
```
